```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from scipy.stats import norm
from non_gaussian_converter import GaussianTransformer
from sklearn.preprocessing import QuantileTransformer
```

```python
e1 = norm(0,1).rvs(200)
e2 = norm(10,1).rvs(200)
x1 = np.hstack((e1,e2))
x2 = np.hstack((e1+10,e2-10)) + norm(0,0.1).rvs(400)
plt.plot(x1)
plt.plot(x2)
```

```python
plt.plot(x1,x2,"o")
```

```python
np.corrcoef(np.hstack((x1[:,None],x2[:,None])).T)
```

```python
# x1t = GaussianTransformer().fit_transform(x1)
# x2t = GaussianTransformer().fit_transform(x2)
x1t = QuantileTransformer(output_distribution="normal").fit_transform(x1.reshape(-1,1))
x2t = QuantileTransformer(output_distribution="normal").fit_transform(x2.reshape(-1,1))
```

```python
plt.plot(x1t)
plt.plot(x2t)
```

```python
plt.plot(x1t,x2t,"o")
```

```python
np.corrcoef(np.hstack((x1t,x2t)).T)
```
